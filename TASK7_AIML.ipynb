```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('breast-cancer.csv')
df.info()
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
svc_linear=SVC(kernel='linear')
svc_linear=svc_linear.fit(x_train_scaled,y_train)
y_pred=svc_linear.predict(x_test_scaled)
print('linear kernel result: ')
print(confusion_matrix(y_test,y_pred))
print('accuracy score is:',accuracy_score(y_test,y_pred))
print('classification report is: ',classification_report(y_test,y_pred))
```

```python
svm_rbf=SVC(kernel='rbf')
svm_rbf.fit(x_train_scaled,y_train)
y_rbf_pred=svm_rbf.predict(x_test_scaled)
print("RBF Kernel Results:")
print(confusion_matrix(y_test, y_rbf_pred))
print(classification_report(y_test, y_rbf_pred))
print("Accuracy:", accuracy_score(y_test, y_rbf_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001],'kernel':['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)
grid.fit(x_train_scaled, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Cross-Val Score (train):", grid.best_score_)
print("Test Accuracy with Best Model:", grid.best_estimator_.score(x_test_scaled, y_test))
```

```python
from sklearn.model_selection import cross_val_score
model = SVC(C=1, gamma=0.01, kernel='rbf')
scores = cross_val_score(model, x_train_scaled, y_train, cv=5)

print("Cross-validation scores:", scores)
print("Average CV score:", scores.mean())
```
